```python
#10) Calculate the median user score and replace NaNs with the median score
#df_games['User_Score'].median()
#This give a cannot convert a string to float error
#Let's determine the string
df_games['User_Score'].unique()
```

From the output above, we see that there is a string 'tbd' in the critic score entries. We will replace this with NaN and then recalulate the median.

```python
#10) continued
#replace 'tbd' with NaN
df_games['User_Score'] = df_games['User_Score'].replace('tbd', np.NaN)
#find the median user score
df_games['User_Score'].median()
```

```python
#10) continued
#replace user score NaNs with median value
df_games['User_Score'].fillna(df_games['User_Score'].median(), inplace = True)
#display the new data frame
df_games.head(5)
```
