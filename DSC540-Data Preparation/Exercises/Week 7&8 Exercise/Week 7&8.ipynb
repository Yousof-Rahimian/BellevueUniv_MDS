# Grouping by Index Levels


```python
#Create a dataset  
columns = pd.MultiIndex.from_arrays([['USA', 'USA', 'USA', 'JP', 'JP', 'IR', 'IR', 'IR'],
                                    [1, 3, 5, 1, 3, 4, 6, 4]],
                                    names=['cty', 'tenor'])
hier_df = pd.DataFrame(np.random.randn(9, 8), columns=columns)
hier_df
```

```python
# Use the 'level' keyword to pass the level number 
hier_df.groupby(level='cty', axis=1).count()
```

# Generating Date Ranges


```python
# Generating a DatetimeIndex '2020-01-03', '2022-01-05'
index = pd.date_range('2020-05-01', '2022-05-01')
index
```

```python
# Indicate start and end date with 40 days period
pd.date_range(start='2020-05-01', periods=40)
pd.date_range(end='2022-05-01', periods=40)
```

```python
# last business day(weekday) of month for 2022
pd.date_range('2022-01-01', '2022-12-01', freq='BM')
```

```python
# Generate a set of timestamps normalized to midnight
pd.date_range('2022-05-01 12:56:31', periods=5, normalize=True)
```
