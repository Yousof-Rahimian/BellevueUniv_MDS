# Activity01: Handling List

Create a list of 100 random numbers 

```python
import random

LIMIT = 100
random_number_list = [random.randint(0, LIMIT) for x in range(0, LIMIT)]
```

```python
random_number_list
```

Create a new list from this random list, which numbers that are divisible by 3.

```python
list_with_divisible_by_3 = [a for a in random_number_list if a % 3 == 0]
```

```python
list_with_divisible_by_3
```

Calculate the length of these two lists and store the differences in a new variable.

```python
length_of_random_list = len(random_number_list)
length_of_3_divisible_list = len(list_with_divisible_by_3)
difference = length_of_random_list - length_of_3_divisible_list
difference
```

Using a loop, perform step2 and 3 and find the diffrence variable three times.

```python
NUMBER_OF_EXPERIEMENTS = 3
difference_list = []
for i in range(0, NUMBER_OF_EXPERIEMENTS):
    random_number_list = [random.randint(0, LIMIT) for x in range(0, LIMIT)]
    list_with_divisible_by_3 = [a for a in random_number_list if a % 3 == 0]
    
    length_of_random_list = len(random_number_list)
    length_of_3_divisible_list = len(list_with_divisible_by_3)
    difference = length_of_random_list - length_of_3_divisible_list
    
    difference_list.append(difference)
difference_list
```

Find the arithmetic mean of these three difference value.

```python
avg_diff = sum(difference_list) / float(len(difference_list))
avg_diff
```

# Activity03: Permutation, Iterator, Lambda, List

```python
from itertools import permutations, dropwhile
```

```python
#Using permutations to to generate all possible three-digit numbers. 
permutations(range(3))
```

```python
#Using isinstance and assert to make sure the the return types are tuples.
for number_tuple in permutations(range(3)):
    print(number_tuple)
    assert isinstance(number_tuple, tuple)
```

```python
#Using dropwhile and lambda exprssions to convert all the tuples to lists. 
for number_tuple in permutations(range(3)):
    print(list(dropwhile(lambda x: x <= 0, number_tuple)))
```

```python
#Returne the actul number contained in the list.
import math
def convert_to_number(number_stack):
    final_number = 0
    for i in range(0, len(number_stack)):
        final_number += (number_stack.pop() * (math.pow(10, i)))
    return final_number
```

```python
for number_tuple in permutations(range(3)):
    number_stack = list(dropwhile(lambda x: x <= 0, number_tuple))
    print(convert_to_number(number_stack))
```

# Activity04: Design your own CSV parser

Import zip_longest

```python
from itertools import zip_longest
```

```python
def return_dict_from_csv_line(header, line):
    # Zip them
    zipped_line = zip_longest(header, line, fillvalue=None)
    # Use dict comprehension to generate the final dict
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict
```

Open the accompanying sales_record.csv by using r mode and start reading the file line by line

```python
with open("sales_record.csv", "r") as fd:
    first_line = fd.readline()
    header = first_line.replace("\n", "").split(",")
    for i, line in enumerate(fd):
        # Here we loop over the first 10 lines in order to not to make the output too big
        line = line.replace("\n", "").split(",")
        d = return_dict_from_csv_line(header, line)
        print(d)
        if i > 10:
            break
```
