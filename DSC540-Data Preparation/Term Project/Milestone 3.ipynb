# Getting the HTML

```python
#Get the url of the webpage we want to scrape
url = 'https://en.m.wikipedia.org/wiki/Electric_car'    
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'lxml')
```

```python
#Get the html of webpage
soup
```

# Getting the Table 

```python
#Get the table tag using soup function 
table_data = soup.find('table', {'class':"wikitable plainrowheaders sortable"})
```

```python
#Get the headres of the tables with create a empy list 
headers = []
```

```python
#for i in table_data.find("tbody").find("tr").findAll("th"):
    #headers.append(i.getText().strip())
```

```python
#print(headers)
```

```python
#Find the html that corresponds to the headers 
#Ask BeautiFul soup to go through the table and find the everythingh with 'th' tag using for loop
for i in table_data.find_all('th'):
    title = i.text.strip()
#Getting all the headers of the table 
    headers.append(title)
```

```python
#Create a DataFrame of headers we got 
df = pd.DataFrame(columns = headers)
```

```python
df
```

```python
#Getting the data within the table
#Using for loop to find all the rows by calling a 'tr'
all_data = []
for row in table_data.select("tr:has(td)"):
    #Find all data in each row with the 'td' tags
    #Get the text of each tds and adding the row back in the data frame
    tds = [td.get_text(strip=True) for td in row.select("td")]
    all_data.append(tds)
#Getting the DataFrame with all data
df = pd.DataFrame(all_data, columns=headers)
df
```

```python
#Identify variables and it's attributes in the dataset
df.shape
df.info()
df.describe()
```
